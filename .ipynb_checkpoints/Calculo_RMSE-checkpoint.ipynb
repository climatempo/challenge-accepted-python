## 2. Carregamento e tratamento dos dados

Os dados são carregados utilizando a biblioteca XArray e convertidos para *Dataframes* do Pandas.

```python
ds = xr.open_dataset('observation.nc')
observation = ds.to_dataframe()
```

```python
ds = xr.open_dataset('forecast.nc')
forecast = ds.to_dataframe()
```

```python
observation = observation.reset_index()
```

```python
forecast = forecast.reset_index()
forecast['t2m'] = (forecast['t2m'] - 273) # Convertendo de Kelvin para Celsius
```

Aqui são removidos alguns valores de temperatura que estão nulos no arquivo *observation*.

```python
forecast = forecast[~observation['temperatura'].isna()]
observation = observation[~observation['temperatura'].isna()]
```

Para faciltar a busca por pontos geográficos no *dataset* são gerados *geohashes* para cada ponto.

```python
observation['geohash'] = observation.apply(lambda row: pgh.encode(row['lat'], row['lon']), axis = 1)
```

```python
forecast['geohash'] = forecast.apply(lambda row: pgh.encode(row['lat'], row['lon']), axis = 1)
```

```python
observation_locations_geohashes = observation['geohash'].unique()
```

Método para 'cortar' o *dataset* em porções de seis horas.

```python
def cut_dataframe(df, index_start, index_end, step):
    dataframes = []
    
    for i in range(index_start, index_end, step):
        end = i + (step - 1)

        if end <= index_end:
            dataframes.append(df.loc[i: end])
        else:
            dataframes.append(df.loc[i: -1])
    
    return dataframes
```

Percorrendo cada ponto (utilizando os *geohashes*), efetuando os cortes no intervalo de seis horas e calculando o RMSE para cada intervalo. Por fim é gerado um *dataframe* com os intervalos para cada ponto.

```python
rmse_df = pd.DataFrame()

for geohash in observation_locations_geohashes:
    point_observation_data = observation[observation['geohash'] == geohash]
    
    index_start = point_observation_data.index[0]
    index_end = point_observation_data.index[-1]
    
    point_forecast_data = forecast.loc[index_start: index_end]
    
    observation_cuts = cut_dataframe(point_observation_data, index_start, index_end, 6)
    forecast_cuts = cut_dataframe(point_forecast_data, index_start, index_end, 6)
    
    data = {"lat": [], "lon": [], "start": [], "end": [], 'rmse': []}
    
    for i in range(0, len(observation_cuts)):
        observation_cut = observation_cuts[i]
        forecast_cut = forecast_cuts[i]
        
        data['lat'].append(observation_cut.iloc[0]['lat'])
        data['lon'].append(observation_cut.iloc[0]['lon'])
        
        data['start'].append(observation_cut.time.min())
        data['end'].append(observation_cut.time.max())
        
        data['rmse'].append(mean_squared_error(observation_cut['temperatura'], forecast_cut['t2m']))
    
    rmse_df = pd.concat([rmse_df, pd.DataFrame(data)])
```

## 4. Mapas e gráfico do ponto -23.5489, -46.6388

### 4.1 Tratamento dos dados

Não há exatamente esse ponto no *dataset*, então é selecionado o ponto mais próximo.

```python
sp = rmse_df.query('lat >= -23.6 and lat <= -23.52').query('lon >= -46.9 and lon <= -46.6')
```

Colocando os dados em um *GeoDataFrame*.

```python
geometries = sp.apply(lambda row: Point(row['lon'], row['lat']), axis = 1)
sp_gdf = gpd.GeoDataFrame(data = sp[['lat', 'lon', 'start', 'rmse']], geometry = gpd.GeoSeries(geometries))
sp_gdf.crs = 'epsg:4326'
```

Carregamento do *shape* (malha) da cidade de São Paulo. Essa malha veio do IBGE.

```python
sp_geometry = gpd.read_file('sp/sp.shp')
sp_geometry.to_crs('epsg:4326', inplace = True)
```

### 4.2 Mapas do RMSE

Desenho dos doze mapas com o RMSE para cada horário em períodos de seis horas.

```python
for row in sp_gdf.iterrows():
    geo = gpd.GeoDataFrame({'rmse': row[1]['rmse'], 'geometry': gpd.GeoSeries(row[1]['geometry'])})
    
    gplt.pointplot(
        geo, hue = 'rmse', legend=True, legend_var='hue',
        extent = sp_geometry.total_bounds,
        ax = gplt.polyplot(sp_geometry)
    )
    
    plt.title("Raiz do erro médio quadrático (RMSE) em " + row[1]['start'].strftime("%d/%m/%Y %H:%M"), fontsize=18)
    plt.figtext(.55, .67, "RMSE: {:.5}".format(str(row[1]['rmse'])), fontsize=12, ha='center')
```

### 4.3 Gráfico do RMSE

Gráfico com o RMSE para cada horário em períodos de seis horas.

```python
chart_df = sp_gdf[['start', 'rmse']]
chart_df.columns = ['ticks', 'rmse']
chart_df = chart_df.reset_index()
chart_df['ticks'] = chart_df.apply(lambda row: row['ticks'].strftime("%d/%m %H:%M"), axis = 1)
```

```python
plt.figure(figsize=(20,5))

plt.plot('ticks', 'rmse', label = 'RMSE', data = chart_df, marker='o', markerfacecolor='white', markersize=6, color='red', linewidth=3)

plt.title('Raiz do erro médio quadrático (RMSE) na região central de São Paulo')
plt.xlabel('Data (ano 2018)')
plt.ylabel('RMSE')
```

## 5. Geração do arquivo NetCDF

Tratamento no *DataFrame* do RMSE, separando as colunas que irão para o arquivo final, renomeando a coluna de horário e definindo o índice.

```python
rmse_netcdf_df = rmse_df[['lat', 'lon', 'start', 'rmse']]
rmse_netcdf_df.columns = ['lat', 'lon', 'time', 'rmse']
rmse_netcdf_df.set_index(['lat', 'lon', 'time'], inplace = True)
```

Transformação do *DataFrame* para o formato do XArray e geração do arquivo NetCDF.

```python
rmse_netcdf = rmse_netcdf_df.to_xarray()
rmse_netcdf.to_netcdf('rmse.nc')
```
